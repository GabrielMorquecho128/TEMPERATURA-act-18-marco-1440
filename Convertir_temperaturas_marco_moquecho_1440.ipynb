## 1. Fórmulas de Conversión

Para convertir entre estas dos escalas, utilizamos relaciones matemáticas basadas en los puntos de congelación y ebullición del agua.

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(O su equivalente decimal: $°F = (°C \times 1.8) + 32$)*

* **De Fahrenheit a Celsius:**
    $$°C = \frac{°F - 32}{1.8}$$
    *(O su equivalente fraccionario: $°C = (°F - 32) \times \frac{5}{9}$)*

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo 1: Convertir 25°C a Fahrenheit (Temperatura ambiente)
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Multiplicar:** $25 \times 1.8 = 45$
4.  **Sumar 32:** $45 + 32 = 77$
* **Resultado:** 25°C equivalen a **77°F**.

### Ejemplo 2: Convertir 98.6°F a Celsius (Temperatura corporal promedio)
1.  **Identificar la fórmula:** $°C = \frac{°F - 32}{1.8}$
2.  **Sustituir el valor:** $°C = \frac{98.6 - 32}{1.8}$
3.  **Restar 32:** $98.6 - 32 = 66.6$
4.  **Dividir entre 1.8:** $66.6 \div 1.8 = 37$
* **Resultado:** 98.6°F equivalen a **37°C**.

## 3. Código en Python

Este programa utiliza funciones específicas y un ciclo interactivo para que puedas realizar tantas conversiones como necesites.

```python
# Función para convertir de Celsius a Fahrenheit
def convertirCF():
    celsius = float(input("\nIngrese los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

# Función para convertir de Fahrenheit a Celsius
def convertirFC():
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

# Función principal que contiene el Menú
def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")
        
        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida, por favor intente de nuevo.")

# Ejecución del programa
if __name__ == "__main__":
    menu()
```

### Explicación del Código:
* **`while True`**: Crea un bucle infinito para que el menú aparezca después de cada conversión. Solo se detiene cuando eliges la opción 3.
* **`float(input(...))`**: Permite que el usuario ingrese números con decimales.
* **`f"{valor:.2f}"`**: Es una forma de darle formato al texto para que solo muestre 2 decimales en el resultado final, lo cual es más limpio y profesional.
* **`if/elif/else`**: Controla la lógica de decisión según el número que el usuario escriba.
